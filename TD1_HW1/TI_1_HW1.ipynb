```python
# Load modules
import numpy as np
import matplotlib.pyplot as plt
```

# Theory on the Gradient Descent algorithm

```python
# Function to be optimized
def f(x, y):
    return (x + 2*y - 7)**2 + (2*x + y - 5)**2 # Booth's function

def gradient(x, y):
    dfdx = 2*(x + 2*y - 7) + 4*(2*x + y - 5)
    dfdy = 4*(x + 2*y - 7) + 2*(2*x + y - 5)
    return np.array([dfdx, dfdy])
```

```python
# Run gradient descent algorithm
lr = 0.01 # Learning rate
iterations = 1000 # Number of iterations
start_point = np.array([0.0, 0.0]) # Initial point

# Store optimization path
path = [start_point.copy()]
current = start_point.copy()

for _ in range(iterations):
    grad = gradient(current[0], current[1])
    current -= lr * grad
    path.append(current.copy())
```

```python

# Convert path to numpy array if it isn't already
path = np.array(path)  # Add this line

# Plot the results
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

# Create figure with two subplots
fig = plt.figure(figsize=(15, 7))

# First subplot
ax1 = fig.add_subplot(121)
ax1.plot(Z)
ax1.scatter(start_point[0], start_point[1], c='r', label='Start point')
ax1.legend()

# Second subplot
ax2 = fig.add_subplot(122)
contour = ax2.contourf(X, Y, Z, levels=50, cmap='viridis')
ax2.plot(path[:,0], path[:,1], 'r.-', markersize=2, linewidth=1, label='Optimization path')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Contour plot with Optimization Path')
plt.colorbar(contour, ax=ax2)

ax2.legend()

plt.colorbar(contour)
plt.tight_layout()
plt.show()
```
