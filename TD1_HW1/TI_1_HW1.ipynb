# Theory on the Gradient Descent algorithm

```python
# Function to be optimized
def f(x, y):
    return (x + 2*y - 7)**2 + (2*x + y - 5)**2 # Booth's function

def gradient(x, y):
    dfdx = 2*(x + 2*y - 7) + 4*(2*x + y - 5)
    dfdy = 4*(x + 2*y - 7) + 2*(2*x + y - 5)
    return np.array([dfdx, dfdy])
```

```python
# Run gradient descent algorithm
lr = 0.01 # Learning rate
iterations = 1000 # Number of iterations
start_point = np.array([0.0, 0.0]) # Initial point

# Store optimization path
path = [start_point.copy()]
current = start_point.copy()

for _ in range(iterations):
    grad = gradient(current[0], current[1])
    current -= lr * grad
    path.append(current.copy())
```

```python
# Plot the results
path = np.array(path)

# Plot the results
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

# Initial plot for the Booths function
plt.figure(figsize=(14, 6))
plt.plot(Z)
plt.title('Z Values Plot')
plt.show()
```

```python
# Contour in 2D
plt.figure(figsize=(14, 6))
contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.plot(path[:,0], path[:,1], 'r.-', markersize=2, linewidth=1, label='Optimization path')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour plot with Optimization Path')
plt.colorbar(contour)
plt.legend()
plt.show()
```

```python
# Second plot - 3D Surface
fig = plt.figure(figsize=(14, 6))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.scatter(path[:,0], path[:,1], f(path[:,0], path[:,1]), c='r', s=20, label='Optimization Path')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot with Optimization Path')
plt.show()
```

# Gradient Descent Methodology

## Geometric Interpretation

The Booth function creates a paraboloid surface in three-dimensional space. When viewed from above, its contour lines form elliptical shapes centered around the global minimum at (1,3). These ellipses aren't perfectly circular due to the cross-terms in the function, which create a slight tilt in the surface. This tilt explains why gradient descent paths often take a curved route rather than moving directly toward the minimum.

## Convergence Analysis

The convergence behavior of gradient descent on the Booth function is particularly interesting due to its quadratic nature. The rate of convergence follows an exponential decay pattern, where the distance to the optimum decreases by a constant factor in each iteration. This factor depends on the ratio of the largest to smallest eigenvalues of the Hessian matrix.
For the Booth function, we can derive the exact convergence rate. The error at iteration k, denoted as e(k), follows:
$$ e(k) ‚â§ (1 - 2Œ∑/L)^k * e(0) $$
where L is the Lipschitz constant of the gradient (equal to the largest eigenvalue of the Hessian).

## Optimization Landscape

The optimization landscape of the Booth function has several interesting properties. The level sets (contours) become more closely spaced as we move away from the minimum, indicating steeper gradients. This means gradient descent typically makes larger steps when far from the optimum and naturally slows down as it approaches the solution.

**Objective Function**: The Booth Function $f(x,y) = (z + 2y - 7)^2 + (2x + y - 5)^2$ is used, which has a global minimum at $(1,3)$

**Algorithm Steps**:

1. **Gradient Computation**:
    The partial derivatives of the function are calculated:

    $$\frac{\partial f}{\partial x} = 2(x + 2y - 7) + 4(2x + y - 5)$$

    $$\frac{\partial f}{\partial y} = 4(x + 2y - 7) + 2(2x + y - 5)$$

2. **Parameter Update**:
    At each iteration, the parameters $(x,y)$ are updated using: 

    $$x_{new} = x_{old} - \eta \cdot \frac{\partial f}{\partial x}$$

    $$y_{new} = y_{old} - \eta \cdot \frac{\partial f}{\partial y}$$

3. **Convergence**:
    A fixed number of iterations $(1000)$ is used for simplicity. The optimization path is tracked to visualize convergence.

**Visualization**:
 - 3D Surface Plot: Shows the function's landscape and the optimization trajectory.
 - Contour Plot: Highlights the path on the function's level sets, illustrating how gradient descent navigates toward the minimum.
